```python
import fitz  # PyMuPDF
from pdf2image import convert_from_path
import pytesseract

# Paths
pdf_path = "Sample Document with Separators (1).pdf" 
poppler_path = r"C:\Users\Vinee\OneDrive\Attachments\Desktop\PROTECT\Release-24.08.0-0\poppler-24.08.0\Library\bin"
separator_keyword = "R 1.0"

# Step 1: Extract embedded text
doc = fitz.open(pdf_path)
embedded_text = [page.get_text() for page in doc]

# Step 2: Convert PDF to images
images = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)

# Step 3: OCR each image
ocr_text = [pytesseract.image_to_string(img) for img in images]

# Step 4: Combine embedded + OCR text for each page
combined_text = [embedded_text[i] + "\n\n" + ocr_text[i] for i in range(len(embedded_text))]

# Step 5: Detect separators
separator_pages = []
for i, text in enumerate(combined_text):
    if separator_keyword.lower() in text.lower():
        separator_pages.append(i + 1)  # human-readable page number

# Output
print("Extraction complete.")
print(f"Pages with separator '{separator_keyword}': {separator_pages}")
```

```python
import os
import fitz  # PyMuPDF

# Input PDF path
pdf_path = "Sample Document with Separators (1).pdf"  # <-- update this to your actual filename
base_filename = os.path.splitext(os.path.basename(pdf_path))[0]

# Load the PDF
doc = fitz.open(pdf_path)
total_pages = doc.page_count

# Add start (0) and end page (total) to the separator list
separator_indices = [0] + separator_pages + [total_pages + 1]

# Loop over separator ranges and save each split
for i in range(1, len(separator_indices)):
    start_page = separator_indices[i - 1]
    end_page = separator_indices[i] - 1  # don't include the separator page

    # Skip empty sections
    if start_page >= end_page:
        continue

    # Create new PDF
    split_doc = fitz.open()
    for page_num in range(start_page, end_page):
        split_doc.insert_pdf(doc, from_page=page_num, to_page=page_num)

    # Save with format: originalname_split#.pdf
    split_filename = f"{base_filename}_split{i}.pdf"
    split_doc.save(split_filename)
    split_doc.close()

    print(f"Saved: {split_filename}")

doc.close()
```

```python
# --- Load the LLM ---
llm = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama3-8b-8192"  # or use "Llama3-8b-8192"
)

# --- Your Priority-Based Classification Logic ---
SORTING_LOGIC = """
You are an intelligent document classification agent for a life insurance company. Your job is to read the text of each split PDF and assign it to a specific ClientID using the following priority-based logic:

1. **PO Box Check** (Highest Priority):
If any of the following PO Box numbers are found, assign accordingly:
- 2869 or 42000 → ClientID: NEWBIZ
- 2873 or 42001 → ClientID: CUSTSERV
- 2875 or 42002 → ClientID: LICCOMM

2. **Form Type Check** (If no PO Box):
If any of the following forms are found, assign accordingly:
- Forms like “1035 Exchange Absolute Agreement” or “Application for Life Insurance” → NEWBIZ
- Forms like “Payment Coupon”, “Ownership Change Request”, “Policy Change Form”, “Beneficiary Change Request”, “Name Change Request” → CUSTSERV
- Forms like “Other Interested Party Authorization Request” → CUSTSERV

3. **Keyword Check** (Lowest Priority):
If keywords like: CWA (Cash with Application), COD (Cash on Delivery), Requirement, Delivery acknowledgment, amendment appear → assign to NEWBIZ

4. **Subpoena Keywords Check** (Override):
If any of these legal keywords are found, immediately assign to: SUBPOENA:
Subpoena, Duces Tecum, Law attorney, Court summons, Command, Commanded, Court appearance, Court Hearing, Court Schedule, Plaintiff, Defendant

Only return the ClientID in your response.
"""

# --- Prompt Template for LangChain ---
prompt = ChatPromptTemplate.from_template(
    """
{sorting_logic}

<context>
{document_text}
</context>

What ClientID does this document belong to?
Only return the ClientID.
"""
)

# --- Chain Together Prompt + LLM ---
chain = prompt | llm

# --- Function: Extract text from PDF using PyMuPDF ---
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text

# --- Classify All Split PDFs in Current Folder ---
import os

for file in split_files:
    doc_text = extract_text_from_pdf(file)
    response = chain.invoke({
        "sorting_logic": SORTING_LOGIC,
        "document_text": doc_text
    })
    print(f" {file} → Assigned ClientID: {response.content}")
```

```python
import os
import traceback
from pdf2image import convert_from_path
import pytesseract
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq

# --- LLM Setup ---
llm = ChatGroq(
    groq_api_key=groq_api_key,
    model_name="llama3-8b-8192"
)

# --- Efficient Sorting Logic ---
SORTING_LOGIC = """
Assign the document to one of the following ClientIDs using this logic:

1. PO Box (highest priority):
- 2869, 42000 → NEWBIZ
- 2873, 42001 → CUSTSERV
- 2875, 42002 → LICCOMM

2. Forms (if no PO Box found):
- “1035 Exchange”, “Life Insurance Application” → NEWBIZ
- “Payment Coupon”, “Ownership Change”, “Policy Change”, “Beneficiary Change”, “Name Change” → CUSTSERV
- “Interested Party Authorization” → CUSTSERV

3. Keywords (if no forms):
- CWA, COD, Requirement, Acknowledgment, Amendment → NEWBIZ

4. Subpoena override:
- If any legal terms like Subpoena, Court, Plaintiff, Defendant → assign to SUBPOENA

Only return the ClientID.
"""

# --- Prompt + Chain ---
prompt = ChatPromptTemplate.from_template(
    """
{sorting_logic}

<context>
{document_text}
</context>

What ClientID does this document belong to?
Return ONLY the ClientID.
"""
)
chain = prompt | llm

# --- OCR Extraction with Size Reduction ---
def extract_text_with_ocr(pdf_path, poppler_path, max_chars=3000):
    try:
        images = convert_from_path(pdf_path, dpi=300, poppler_path=poppler_path)
        text = ""
        for img in images:
            if len(text) > max_chars:
                break
            text += pytesseract.image_to_string(img)
        cleaned = " ".join(text.split())  # remove newlines, extra spaces
        return cleaned[:max_chars]
    except Exception as e:
        raise RuntimeError(f"OCR failed for {pdf_path}: {str(e)}")

# --- Set Poppler Path ---
poppler_path = r"C:\Users\Vinee\OneDrive\Attachments\Desktop\PROTECT\Release-24.08.0-0\poppler-24.08.0\Library\bin"

# --- Get All PDF Files ---
split_files = [f for f in os.listdir(".") if f.endswith(".pdf")]

# --- Process Each PDF ---
for file in split_files:
    print(f"\nProcessing: {file}")
    try:
        doc_text = extract_text_with_ocr(file, poppler_path)

        if not doc_text.strip():
            raise ValueError("Extracted text is empty. Skipping file.")

        response = chain.invoke({
            "sorting_logic": SORTING_LOGIC,
            "document_text": doc_text
        })

        print(f"✅ {file} → Assigned ClientID: {response.content.strip()}")

    except Exception as err:
        print(f"❌ Error processing file: {file}")
        print("Error:", str(err))
        traceback.print_exc()
```
